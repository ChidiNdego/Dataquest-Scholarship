Project concentrates on exploring data on job outcomes with visualizing and extracting insightful information about the data.

```python
#importing the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#reading the data into pandas library
recent_grads = pd.read_csv('recent-grads.csv')
```

### Description of data column names
* Rank - Rank by median earnings (the dataset is ordered by this column).
* Major_code - Major code.
* Major - Major description.
* Major_category - Category of major.
* Total - Total number of people with major.
* Sample_size - Sample size (unweighted) of full-time.
* Men - Male graduates.
* Women - Female graduates.
* ShareWomen - Women as share of total.
* Employed - Number employed.
* Median - Median salary of full-time, year-round workers.
* Low_wage_jobs - Number in low-wage service jobs.
* Full_time - Number employed 35 hours or more.
* Part_time - Number employed less than 35 hours.

Clearly, male graduates largely dominate their female counterparts in Computers & Mathematics and Engineering categories. The females, on the other hand, dominate in Health, Education, Communications & Journalism, Humanities & Liberal Arts, and Psychology & Social Work categories. All other categories have considerably close number of male and female graduates.
